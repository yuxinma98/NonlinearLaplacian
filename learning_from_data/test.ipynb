```python
import torch
accuracy = 0
for data in planted:
    A,y = data
    evals = np.linalg.eigvalsh(A)
    top_eval = evals[-1]
    print(top_eval)
    pred = top_eval > 2.01
    accuracy += pred == y
print(accuracy/len(planted))
```
