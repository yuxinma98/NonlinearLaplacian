```python
import sys
import os

module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)

from data import NonnegativePCADataset, NonnegativePCARecoveryDataset
import numpy as np
```

```python
planted = NonnegativePCADataset(100, 2000, 1.3, "train")
```

```python
import torch

accuracy = 0
for data in planted:
    A, y = data
    evals = np.linalg.eigvalsh(A)
    top_eval = evals[-1]
    print(top_eval)
    pred = top_eval > 2.01
    accuracy += pred == y
print(accuracy / len(planted))
```

```python
recovery = NonnegativePCARecoveryDataset(100, 2000, 1.3, "train")
for data in recovery:
    A, y = data
    evals, evecs = np.linalg.eigh(A)
    top_evecs = evecs[:, -1]
    top_eval = evals[-1]
    print(top_eval, (top_evecs * np.array(y)).sum() ** 2)
```
